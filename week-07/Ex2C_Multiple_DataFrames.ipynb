# Exercise 2.C Working with multiple DataFrames

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Basic Aggregations

```python
data = {'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],    # creating DataFrame 
 'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East',
'West', 'North'],
 'Units_Sold': [10, 15, 20, 5, 10, 15, 10, 5, 15],
 'Price': [100, 200, 150, 100, 200, 150, 100, 200, 150]}
df_sales = pd.DataFrame(data)
df_sales
```

```python
df_sales["Revenue"] = df_sales["Units_Sold"] * df_sales["Price"] # to creat a column name Revenue by multiplying units_sold and price
df_sales
```

```python
df_sales[["Units_Sold", "Revenue"]].sum()  # total number of units sold and total revenue
```

```python
df_sales[["Units_Sold", "Revenue"]].mean() # avg number of units sold and avg revenue
```

## Grouping Data

```python
df_sales.groupby("Product")[["Units_Sold", "Revenue"]].sum() # grouped by product(unique values) and
                                                            # calculated the total Units_Sold and total Revenue for each product
```

```python
df_sales.groupby("Region")[["Units_Sold", "Revenue"]].mean() # grouped by Region(unique values) and
                                                             # calculated the avg Units_Sold and total Revenue for each product
```

```python
df_sales.groupby(["Product", "Region"])[["Units_Sold", "Revenue"]].sum() # grouped them by product and region and 
                                                                        # # calculated the total Units_Sold and total Revenue for the combination
```

## Sorting and Ranking Data

```python
df_sales.sort_values("Units_Sold") # sorting by units sold in ascending order
```

```python
df_sales.sort_values("Revenue", ascending=False) # sorting by revenue in descending order
```

```python
df_sales["Revenue_Rank"] = df_sales["Revenue"].rank(ascending=False) # Added a new column "Revenue_Rank in descending order
df_sales
```

## Merging DataFrames

```python
product_data = {'Product': ['A', 'B', 'C'],
 'Category': ['Electronics', 'Furniture', 'Appliances']}
df_products = pd.DataFrame(product_data)
```

```python
df_merged = pd.merge(df_sales, df_products, on="Product") # merged the existed data(df_sales) and the new data(df_products)
df_merged
```

```python
df_merged.groupby("Category")["Revenue"].sum() # grouped by category and total revenue by category
```

## Bonus creating a custom chart

```python
category_revenue = df_merged.groupby("Category")["Revenue"].sum()
```

```python
category_revenue.plot.bar(color='purple')
plt.title('Total Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.show()
```
