## Permutations

```python
print(r_4)
```

```python
np.random.shuffle(r_4) # The rows are shuffled but the values inside the rows are not shuffled and it shuffles the original.
print(r_4)
```

```python
new_shuffle = np.random.permutation(r_4) # Permutation creats a new shuffeled copy. 
print(new_shuffle)
print(r_4)
```

I would use shuffle to randomly shuffle the original and I would use permutation when I want to keep the original as it is but shuffle a copy.

## Cumulative Operations

```python
print(r_4)
```

```python
print(np.sum(r_4)) # sum of all the values
```

```python
print(np.sum(r_4, axis=0)) # adds across the columns
```

```python
print(np.cumsum(r_4)) # it is the rolling value
```

## Grocery Problem


```python
bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])
print(bag1)
print(bag2)
```

```python
print("The first bag contains", np.sum(bag1), "items.")
print("The second bag contains", np.sum(bag2), "items.")
```

```python
total = bag1 + bag2
print("There are", total[0], "vegetables,", total[1], "fruits,", total[2], 
      "boxes of snacks, and", total[3], "beverages in the two grocery bags.")
```

 ## Teamwork

```python
ali = np.array([5, 4, 6, 4, 2])
ben = np.array([0, 5, 5, 3, 4])
carla = np.array([3, 6, 0, 3, 4])
```

```python
print(ali)
print(ben)
print(carla)
```

```python
all_knocks = np.array([ali, ben, carla])
print(all_knocks)
```

```python
print(np.sum(all_knocks, axis=1)) # the total number of doors that each person knocked 
```

```python
print(np.sum(all_knocks)) #  total number of doors that the group collectively knocked on
```

```python
total_by_day = np.sum(all_knocks, axis=0) # how many days were knocked everyday
print(total_by_day)
```

```python
rolling_total = np.cumsum(total_by_day) # the rolling total of knocks
print(rolling_total)
```

```python
estimated_votes = rolling_total * 0.5 # *.5 because 1 vote for every 2 knocks 
print(estimated_votes)
```
