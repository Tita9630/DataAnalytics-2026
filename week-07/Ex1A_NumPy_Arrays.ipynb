# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

## Looping Through Arrays

```python
arrA = np.array([[11, 12, 13],
                [14, 15, 16]])
print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

### The first loop multiplies each row by two and the second loop prints each value separately.

```python
for x in arrA:
    for y in x:
        loop_result = y * 2
        print(loop_result)
```

```python
# The result is the same. The difference is that the multiplied value is stored in a variable before printing.
```

## Joining and Stacking Arrays

```python
joined_array_1 = np.concatenate((arrA,  arrA), axis=0)
print(joined_array_1)
```

```python
joined_array_2 = np.concatenate((arrA,  arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)
```

```python
stacked_array_2 = np.stack((joined_array_1, joined_array_2)) # ValueError: all input arrays must have the same shape.
                                                            # It is because the arrays don't have same shape.
print(stacked_array_2)
```

```python
another_joined_array = np.concatenate((joined_array_1, arrA))
print(another_joined_array)
```

```python
another_2 = np.concatenate((joined_array_1, arrA), axis=1) # The arrays have different numbers of rows. 
print(another_2)                                           # joined_array_1 has 4 rows and arrA has 2 rows.
print(another_2)                                           
```

```python
arrAHS = np.hstack((arrA, arrA)) # it is stacking the arrays horizontaly 
print(arrAHS)
```

```python
arrAVS = np.vstack((arrA, arrA)) # it is stacking the arrays vertically
print(arrAVS)
```

```python
arrADS = np.dstack((arrA, arrA)) # it is stacking the arrays depth-wise
print(arrADS)
```

## Splitting Arrays

```python
splitsky = np.split(joined_array_1, 2) # it has 2 arrays with two dimensions each. 
print(splitsky)
```

```python
spliterino = np.array_split(joined_array_1, 3) # it has 2 dimensional arrays. The first array has two rows
print(spliterino                               # and the rest have one row each because 4 can't be divided evenly into 3.
```

## Searching and Replacing Array Values

```python
result_1 = np.where(arrA % 2 == 0, arrA, "odd")
print(result_1)
```

## Sorting and Filtering Arrays

```python
arrB = (((7,6), (13,5), (7,3)))
arrB_sorted = np.sort(arrB)
print(arrB_sorted)
```

```python
arrB = np.array(((7,6), (13,5), (7,3)))
print(arrB)
z = arrB < 7
arrC = arrB[z]
print(arrC)
```

```python
arrB = np.array(((7,6), (13,5), (7,3)))
print(arrB)
y = arrB >= 7
arrD = arrB[y]
print(arrD)
```
