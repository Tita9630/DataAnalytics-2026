 ## Filtering a DataFrame

```python
df[df["Age"] > 18] # to filter age values > 18
```

```python
df[(df["Score"] > 50) & (df["Score"] < 100)] # to filter score values > 50 and < 100
```

 ## Modifying a DataFrame

```python
df_cleaned = df.drop(df[df["Score"] < 0].index) # to drop a row it uses row with .index
df_cleaned
```

```python
df_cleaned = df_cleaned.drop("City", axis=1) # removing column city. It uses column name to be dropped and axis=1
df_cleaned
```

```python
df_cleaned["Age"] = df_cleaned["Age"].fillna(df_cleaned["Age"].mean()) # replace missing age values with the mean age
df_cleaned
```

```python
df_cleaned["Score"] = df_cleaned["Score"].replace(0, df_cleaned["Score"].mean()) # replace score value of 0 with the mean score
df_cleaned
```

```python
df_cleaned = df_cleaned.dropna()  # remove remaining rows with NaN values
df_cleaned
```

```python
df_cleaned = df_cleaned.set_index("ID") # to remove the column ID and apply the values from ID as index labels
df_cleaned
```
