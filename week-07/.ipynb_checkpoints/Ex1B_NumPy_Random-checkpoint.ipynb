# Exercise 1.B NumPy Random

```python
import numpy as np
```

## Generating Random Numbers

```python
r_1 = np.random.randint(1, 20, 2) # randint to generate integers
print(r_1)                        # specifies the row
```

```python
r_2 = np.random.randint(1, 20, (1, 2)) # randint to generate integer
print(r_2)                             # specifies rows and columns
```

```python
r_3 = np.random.rand(2, 1) # rand to generate float
print(r_3)                 # specifies 2 rows and 1 column
```

```python
r_4 = np.random.randint(20, 100, (4, 5)) # random integers between 
print(r_4)                               # 20, 100 and 4 rows and 5 colums 
```

```python
r_5 = np.random.rand(4, 5) # floats, 4 rows and 5 colums
print(r_5)
```

 ## Generating Random Choices from an Array

```python
students = np.array([
    "Alexus", "Alondra", "Betelehem", "Blake", 
    "Charles", "Corey", "Dagmawi", "Danait",
    "Daniel", "Hana", "Lesley", "Lorah"
])
print(students)
```

```python
call_on = np.random.choice(students)
print(call_on)
```

```python
breakout_groups = np.random.choice(students, (3, 4))
print(breakout_groups) # It wouldn't be an effective way 
                       # to assign breakout groups because 
                       # some of the names assigned more than once. 
                       
```

```python
call_on_weighted = np.random.choice(students, 5,
        p =[0.05, 0.3, 0.1, 0, 0.1, 0.05, 0.05, 0.05, 0.1, 0.05, 0.1, 0.05])
print(call_on_weighted) # Alondra has 30% probability and Blake 
                        # had 0% probability. The other students had
                        # similar probability of being selected. some 0.05 and some 0.1
                        
```

 ## Generating a Normal Distribution

```python
bell_curve = np.random.normal(10, 5, (4, 5))
print(bell_curve)
```
