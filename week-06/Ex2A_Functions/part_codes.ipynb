## Lab 2. Parsing strings – part codes

```python
code = "ACME:123-L"
def get_supplier_code(code): # defining a function
    colon = code.find(":") # to find colon in the code
    return code[:colon] # to return everything before the colon and not included the colon 
print(get_supplier_code(code))
```

```python
def get_product_num(code):
    colon = code.find(":")
    dash = code.find("-") #to find the dash in the code
    return code[colon + 1 :dash] # return what's next to colon and upto but not included dash
print(get_product_num(code))
```

```python
def get_size(code):
    dash = code.find("-")
    return code[dash +1:] # return what's next to dash and everything agter the dash
print(get_size(code))
```

```python
code = "DI:12345-M"
def get_supplier_code(code):
    colon = code.find(":")
    return code[:colon]
print(get_supplier_code(code))
```

```python
def get_product_num(code):
    dash = code.find("-")
    colon = code.find(":")
    return code[colon + 1:dash]
print(get_product_num(code))
```
