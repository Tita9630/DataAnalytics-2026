## Lab 1. Working with names – parsing strings

```python
test_name = 'John Doe'
```

```python
def parse_and_display(name):
 for i in range(len(name)):
  if name[i] == ' ':
   first = name[:i]
   last = name[i+1:]
   print(f"Full name: {name}")
   print(f"First name: {first}")
   print(f"Last name: {last}")
```

```python
parse_and_display(test_name)
```

```python
test_list = ['Grace Flores', 'JB Oinonen', 'Ken Penn']
```

```python
parse_and_display(test_list[0]) # output for Grace Flores
```

```python
parse_and_display(test_list[1]) # output for JB Oinonen
```

```python
parse_and_display(test_list[2]) # output for Ken Penn
```

### Modified version of my function

```python
def parse_and_display2(name):
    first_space = name.find(" ") # to find the first space
    last_space = name.rfind(" ") # to find the second space
    if first_space == -1:
        print("First name:", name)
    elif first_space == last_space:
        first = name[:first_space]
        last = name[first_space + 1:]
        print("First name:", first)
        print("Last name:", last)
    else:
        first = name[:first_space]
        middle = name[first_space + 1:last_space]
        last = name[last_space + 1:]
        print("First name:", first)
        print("Middle name:", middle)
        print("Last name:", last)

    
```

```python
parse_and_display2("Lorde")
```

```python
parse_and_display2("Billie Eilish")
```

```python
parse_and_display2("Megan Thee Stallion")
```
