# Lab 2 Exercise 2.B, working with datetime functions

```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
```

```python
print(today.strftime("Today is %A, %B %d, %Y"))
```

# My birthday

```python
my_birthday = datetime.date(1996, 1, 16)
```

```python
print(my_birthday.strftime("My birthday is %m/%d/%Y"))
```

# Next 90 Days

```python
 ninety_d = today + datetime.timedelta(days=90)
```

```python
print(ninety_d.strftime("90 days from today is %m, %d, %Y"))
```

# Time for Dinner

```python
 dinner_time = datetime.time(18, 30)
```

```python
print(dinner_time.strftime("Let's meet for dinner at %I:%M %p"))
```
