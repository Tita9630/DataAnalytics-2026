# Lab 1 Exercise 2.C, working with defining custom functions

```python
def favorite_things(name, movie, music):
    return f"My name is {name}\nMy favorite movie is {movie}\nMy favorite band is {music}"
```

```python
print(favorite_things("Tihitina", "Ted", "Elevation worship"))
```

```python
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}"
```

```python
print(favorite_place("Switzerland", "it is so beautiful"))
```

```python
print(favorite_place("Ethiopia", "my families are there"))
```
