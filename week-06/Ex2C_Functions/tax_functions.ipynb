# Exercise 2 Lab 3: tax_functions

```python
def get_soc_sec_tax(gross_pay):
    return gross_pay * 0.062
```

```python
def get_medicare_tax(gross_pay):
    return gross_pay * 0.0145
    
```

```python
def get_federal_tax(gross_pay, withholding_code):
    if withholding_code == 0:
        tax_rate = 0.23
    elif withholding_code == 1:
        tax_rate = 0.21
    elif withholding_code == 2:
        tax_rate = 0.195
    elif withholding_code == 3:
        tax_rate = 0.185
    else:
        tax_rate = 0.18 - ((withholding_code - 4) * 0.005)
    return gross_pay * tax_rate

    
```

```python
gross_pay = 750
```

```python
withholding_code = 0
```

```python
print(get_soc_sec_tax(gross_pay))
```

```python
print(get_medicare_tax(gross_pay))
```

```python
print(get_federal_tax(gross_pay, withholding_code))
```
